# EDA for Processed Recipes Dataset


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Explore Reviews Dataset

```python
# Load preprocessed reviews data
reviews_df = pd.read_csv('../data/processed_reviews.csv')
reviews_df.head()
```

```python
reviews_df.columns
```

```python
reviews_df.info()
reviews_df.describe()
```

```python
pd.crosstab(reviews_df['UserEngagementLevel'], reviews_df['IsHealthConsciousUser']).plot(kind='bar', stacked=True)
```

```python
reviews_df['SentimentPolarity'].hist(bins=50)
plt.title("Sentiment Polarity")
```

```python
reviews_df['SentimentCategory'].value_counts().plot(kind='bar')
```

```python
sns.scatterplot(data=reviews_df, x='TastePerceptionScore', y='HealthPerceptionScore', hue='SentimentCategory')
```

```python
reviews_df['HealthyRecipePositiveReview'].value_counts(normalize=True).plot(kind='pie', autopct='%1.1f%%', ylabel='Healthy Recipe Positive Review (%)')
plt.title("Healthy Recipe Positive Review Distribution")
```

```python
reviews_df['ReviewDate'] = pd.to_datetime(reviews_df['ReviewDate'], errors='coerce')
reviews_df['YearMonth'] = reviews_df['ReviewDate'].dt.to_period('M')
reviews_df.groupby('YearMonth')['IsHealthConsciousUser'].mean().plot(xlabel='Year', ylabel='Percentage of Health-Conscious Users (%)', title='Share of Health-Conscious Users Over Time')
plt.title("Percentage of Health-Conscious Users Over Time")
```
